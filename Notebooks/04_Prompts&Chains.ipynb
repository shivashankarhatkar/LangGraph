## **Prompts in langgraph**
Already covered in LangChain.

```python
from langchain_google_genai import ChatGoogleGenerativeAI

from dotenv import load_dotenv
import os
load_dotenv()

if os.environ['GOOGLE_API_KEY']:
    print("GEMINI API Key is set.")
else:
    raise ValueError("OpenAI API Key is not set.")
```

```python
from langchain_core.messages import HumanMessage
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
```

```python
from langchain_core.prompts import ChatPromptTemplate


my_prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant that helps to create and curate posts for a social media platform."),
        ("human","I want to create a post about {topic}")
    ]
)
```

### **Chain With Prompts**

```python
chain = my_prompt | llm
#chain.invoke({"topic": "The importance of data privacy in the digital age"})
```
